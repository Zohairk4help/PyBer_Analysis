```python
# 2 step 2: Import dependencies.
import matplotlib.pyplot as plt
```

```python
# step 3
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# step 4: Create the plot
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

plt.show()
```

```python
# Create the plot with ax.plt() using the quick object-oriented interface method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="gold", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.show()
```

```python
# Create the plot with ax.plt() using the quick object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
plt.plot(x_axis, y_axis)

# Create the plot and add a label for the legend.
ax.plot(x_axis, y_axis, label='Boston', color='green', linewidth=2, marker='*')
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.grid()
```

```python
# 3 create bar chart using the same data points as above from line chart.
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
# Create the plot
plt.bar(x_axis, y_axis)
# Create the plot. no anootation of line width and marker necessary for bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot. by inverting the axis 
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
# Create the plot
plt.barh(x_axis, y_axis)
# Create the plot. no anootation of line width and marker necessary for bar chart
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare ($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# create the scatter plot by using plt.plot() and inside the brackets, we have "O"
plt.plot(x_axis, y_axis, 'o')
```

```python
# create the scatter plot using the following code:
plt.scatter(y_axis, x_axis, c='red', label='Chicago')
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()
plt.title("PyBer Fare per Month")
plt.xlim(0,50)
```

```python
# create the bubble chart using matplotlib method
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# create the bubble chart using OOI method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], c='xkcd:sky blue', label='Boston',edgecolor="Black", linewidth=2)
ax.legend()

ax.set_title("PyBer Fare by Month")
plt.gca().invert_yaxis()
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
plt.xlim(0,50)
```

```python
# create a pie chart using Matplotlib method
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(10, 10))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
